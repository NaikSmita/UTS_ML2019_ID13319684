CONTENT:

The research was conducted to develop a Face recognition system which is insensitive to high variation of light and facial appearance. The paper focused variation in lightning which includes intensity, direction and number of light sources.  Adopting an example of pattern classification approach, they have considered every pixel as a coordinate in a high dimensional space.  They have used the perception considering the face is Lambertian surface without shadowing and observed that every image of face with fixed pose in varying illumination which is lying in the 3D linear surface. However, the same person with different facial expression and appearance when light source illuminates the face from different directions and can produce self-shadowing.  Since the face are not true Lambertian surface, images will deviate from the linear surface. 
So, the approaches described in the paper has two observation:
1.  Images of Lambertian surface in varying illumination with fixed viewpoint lie in 3D high dimensional image space.
2. Because of the region, there exists shadowing, specularities and different facial        expression gives huge deviation in image to image results in image unsuitable for recognition.
 Based on which observations, reached a significantly lower dimensional feature space which excludes both light direction and facial expression. The research depends upon the projection method in which projection directions that are orthogonal to the within class scatter while maintaining discriminability. The algorithm Fisherfaces a derivative of Fisher’s Linear Discriminant (FLD).  The proposed method maximises the ratio of between -class to the within class scatter. FLD is a classical technique to achieve pattern recognition, developed by Robert Fisher which has been applied in four different ways in computer vision and face recognition. The proposed another technique Eigenface technique which has similar computational requirements and this technique is based on linear projecting image on low dimensional subspace. But this Eigenface method uses Principal Component Analysis (PCA) for dimensionality reduction.

2.Eigenface is commonly used method for pattern recognition over correlation which requires large storage and computationally very expensive. The increase of within class scatter is later ignored to retain PCA technique in Eigenface.
